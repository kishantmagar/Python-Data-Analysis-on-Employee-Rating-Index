```python
import pandas as pd
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Raw Data

```python
Employee_Index_df = pd.read_csv('Employee Index.csv')
print(Employee_Index_df)
```

#### Generating EDA report of Employee Index raw Dataset

```python
Employee_Index_profile = ProfileReport(Employee_Index_df)
Employee_Index_profile.to_file(output_file="Employee Index EDA Report.html")
```

#### Cleaned Data

```python
Employee_Index_Cleaned_df = pd.read_csv('Employee Index Cleaned.csv')
print(Employee_Index_Cleaned_df)
```

#### Generating EDA report of Employee Index Cleaned Dataset

```python
Employee_Index_Cleaned_profile = ProfileReport(Employee_Index_Cleaned_df)
Employee_Index_Cleaned_profile.to_file(output_file="Employee Index Cleaned EDA Report.html")
```

```python
data = pd.read_csv('Employee Index Cleaned.csv')

# Calculating average employee rating by age group
employee_rating_by_age_group = data.groupby('age group')['employee_rating'].mean()

# Calculating average employee rating by salary
employee_rating_by_salary = data.groupby('salary')['employee_rating'].mean()

# Plotting the average employee_rating by age group
plt.figure(figsize=(10, 6))
employee_rating_by_age_group.plot(kind='bar', color='blue')
plt.title('Average Employee Rating by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Employee Rating')
for i, employee_rating in enumerate(employee_rating_by_age_group):
    plt.text(i, employee_rating + 0.025, f'{employee_rating:.2f}', ha='center', va='bottom')
plt.show()

# Plotting the average employee rating by salary
plt.figure(figsize=(10, 6))
employee_rating_by_salary.plot(kind='bar', color='green')
plt.title('Average Employee Rating by Salary')
plt.xlabel('Salary')
plt.ylabel('Average Employee Rating')
for i, employee_rating in enumerate(employee_rating_by_salary):
    plt.text(i, employee_rating + 0.025, f'{employee_rating:.2f}', ha='center', va='bottom')
plt.show()
```
